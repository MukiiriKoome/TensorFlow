## Pre-Processing Data

```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

Example data:
- An experimental drug was tested on individuals from ages 13 - 100
- The trial had 2100 participants. Half under 65 years. Other half were over 65 years old.
- 95% of patients 65 or older experienced side effects.
- 95% of patients under 65 experienced no side effects.

```python
for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)

    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)

for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)

    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
```

```python
# Print raw data
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

```python
# Simple array split (without stratification)
split_idx = int(0.8 * len(train_samples))
train_samples, test_samples = train_samples[:split_idx], train_samples[split_idx:]
train_labels, test_labels = train_labels[:split_idx], train_labels[split_idx:]
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform((train_samples).reshape(-1,1))
scaled_test_samples = scaler.fit_transform((test_samples).reshape(-1,1))
```

```python
# Print scaled train data
for i in scaled_train_samples:
    print(i)
```

```python
# Print scaled test data
for i in scaled_test_samples:
    print(i)
```

```python

# Save the array to a file
np.save('scaled_train_samples.npy', scaled_train_samples)  # Saves as a .npy file
```

```python
# Save the array to a file
np.save('train_labels.npy', train_labels)  # Saves as a .npy file
```

```python
# Save the array to a file
np.save('scaled_test_samples.npy', scaled_test_samples)  # Saves as a .npy file
```

```python
# Save the array to a file
np.save('train_samples.npy', train_labels)  # Saves as a .npy file
```
